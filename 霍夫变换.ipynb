## 霍夫变换（Hough Transform）步骤总结

1. 边缘检测  
   对原始图像进行边缘检测（如 Canny 算子），提取可能属于直线的边缘点。

2. 直线参数化  
   使用极坐标形式表示直线：
   $$\rho = x\cos\theta + y\sin\theta$$
   其中：
   - $\rho$ 表示直线到原点的最短距离  
   - $\theta$ 表示该距离方向与 $x$ 轴的夹角  

3. 构建参数空间累加器  
   在 $(\rho,\theta)$ 参数空间中建立二维累加器数组。

4. 投票过程  
   对每一个边缘点 $(x,y)$，遍历不同的 $\theta$，计算对应的 $\rho$，
   并在累加器中对相应的 $(\rho,\theta)$ 位置加一票。

5. 峰值检测  
   在累加器中寻找票数较大的峰值点。

6. 结果映射  
   将检测到的 $(\rho,\theta)$ 参数映射回图像空间，得到对应的直线。
