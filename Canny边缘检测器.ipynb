# Canny边缘检测步骤
## 1. 高斯滤波
**目的**：消除图像中的噪声，为后续边缘检测做预处理。

## 2. 计算梯度
**目的**：寻找图像中梯度变化最大的位置（即边缘区域），捕捉像素值突变的地方。

## 3. 非极大值抑制
**目的**：仅保留梯度方向上的最大值，让检测到的边缘只有 **1 像素宽**。

## 4. 双阈值筛选
### 阈值设定（假设）
- 高阈值 TH = 150
- 低阈值 TL = 50

### 梯度值判断规则
| 梯度值 | 判定结果 |
|--------|----------|
| 200    | 强边缘   |
| 120    | 弱边缘   |
| 60     | 弱边缘   |
| 20     | 噪声     |

**核心逻辑**：
- 梯度值 > 高阈值（TH）→ 强边缘（必保留）
- 低阈值（TL）< 梯度值 ≤ 高阈值（TH）→ 弱边缘（需进一步判断）
- 梯度值 ≤ 低阈值（TL）→ 噪声（直接剔除）

## 5. 边缘连接（滞后阈值）
**目的**：区分有效弱边缘和噪声型弱边缘，完成最终边缘筛选。

### 判断规则
- 弱边缘若与**强边缘相连** → 保留（判定为有效边缘）
- 弱边缘若**孤立存在** → 删除（判定为噪声）

### 示例
- 有效边缘：强边缘 —— 弱边缘 —— 弱边缘（完整边缘，全部保留）
- 无效边缘：孤立弱边缘（大概率噪声，直接删除）

### 总结
1. Canny边缘检测核心流程：高斯滤波去噪 → 计算梯度找边缘 → 非极大值抑制细化边缘 → 双阈值初筛 → 边缘连接最终筛选。
2. 双阈值（TH=150、TL=50）是区分强/弱边缘和噪声的关键，边缘连接通过强边缘关联有效弱边缘。
3. 非极大值抑制是实现“1像素宽边缘”的核心步骤，保证边缘精细度。
